```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

data = pd.read_excel('Nilai_Siswa.xlsx', header=3)
data
```

```python
NA_minimum = data.loc[data['NA'].idxmin()]
NA_minimum
```

```python
NA_maximum = data.loc[data['NA'].idxmax()]
NA_maximum
```

```python
avg_all = data['NA'].mean()
avg_all
```

```python
NH_counts = data['NH'].value_counts()
NH_counts
```

```python
avg_NA_per_kelas = data.groupby('Kelas')['NA'].mean()
avg_NA_per_kelas
```

```python
max_NA_per_kelas = data.loc[data.groupby('Kelas')['NA'].idxmax()]
max_NA_per_kelas
```

```python
min_NA_per_kelas = data.loc[data.groupby('Kelas')['NA'].idxmin()]
min_NA_per_kelas
```

```python
class_averages = data.groupby('Kelas')['NA'].mean()

plt.figure(figsize=(10,6))
avg_NA_per_kelas.plot(kind='bar')
plt.xlabel('Kelas')
plt.ylabel('Nilai rata-rata NA')
plt.title('Rata-rata nilai NA per kelas')

plt.show()

highest_average_class = class_averages.idxmax()
lowest_averages_class = class_averages.idxmin()

print('Kelas dengan nilai NA rata-rata tertinggi adalah kelas', highest_average_class)
print('Kelas dengan nilai NA rata-rata terendah adalah kelas', lowest_averages_class)
```
